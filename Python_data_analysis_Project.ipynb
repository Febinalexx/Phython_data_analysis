# Depression Dataset


This dataset contains information on individuals with various attributes related to their personal and lifestyle factors. It is designed to facilitate analysis in areas such as health, lifestyle, and socio-economic status

Features

- **Name**: The name of the individual.
- **Age**: The age of the individual.
- **Marital Status**: The marital status of the individual. (Single, Married, Divorced, and Widowed)
- **Education Level**:  High School, Associate Degree, Bachelor's Degree, Master's Degree, and PhD.
- **Number of Children**
- **Smoking Status**: Indicates whether the individual is a smoker or not. Possible values are Smoker, Former, and Non-smoker.
- **Physical Activity Level**: The level of physical activity undertaken by the individual. Possible values include Sedentary, Moderate, and Active.
- **Employment Status**: The employment status ( Employed and Unemployed).
- **Income**: The individual's annual income in USD.
- **Alcohol Consumption**: The level of alcohol consumption (Low, Moderate, and High).
- **Dietary Habits**: The dietary habits of the individual (Healthy, Moderate, and unhealthy).
- **Sleep Patterns**: The quality of sleep (Good, Fair, and Poor).
- **History of Mental Illness**: Whether the individual has a history of mental illness. Possible values are Yes and No.
- **History of Substance Abuse**: Whether the individual has a history of substance abuse. Possible values are Yes and No.
- **Family History of Depression**: Indicates if there is a family history of depression. Possible values are Yes and No.
- **Chronic Medical Conditions**: Whether the individual has chronic medical conditions. Possible values are Yes and No.

```python
health_factors = ('History of Mental Illness', 'History of Substance Abuse', 'Family History of Depression')
for factor in health_factors:
    pivot_table = pd.pivot_table(data, values='Chronic Medical Conditions', index='Employment Status', columns=factor, aggfunc='sum')
    
    pivot_table.plot(kind='bar', figsize=(8, 6))
    plt.title(f'Chronic Medical Conditions by Employment Status and {factor}')
    plt.xlabel('Employment Status')
    plt.ylabel('Chronic Medical Conditions')
    plt.xticks(rotation=0)
    plt.legend(title=factor)
    plt.show()
```
